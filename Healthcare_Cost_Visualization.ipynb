```python
import pandas as pd
import numpy as np
```

```python
# Set a seed for reproducibility
np.random.seed(42)
```

```python
# Generate simulated data
categories = ['Hospitalization', 'Medication', 'Doctor Visits', 'Tests', 'Others']
num_categories = len(categories)
```

```python
# Generate random costs for each category
costs = np.random.randint(100, 1000, size=num_categories)

# Create a DataFrame
df = pd.DataFrame({'Categories': categories, 'Costs': costs})

# Save the DataFrame to a CSV file
df.to_csv('healthcare_costs_singapore.csv', index=False)
```

```python
# Display the generated data
print("Simulated Healthcare Cost Data:")
print(df)
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Costs'], labels=df['Categories'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Healthcare Costs in Singapore by Category')
plt.show()
```

```python
# Plot a bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x='Categories', y='Costs', data=df, palette='viridis')
plt.title('Healthcare Costs in Singapore by Category')
plt.xlabel('Categories')
plt.ylabel('Costs (in SGD)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
